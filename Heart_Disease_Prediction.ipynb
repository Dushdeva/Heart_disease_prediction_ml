```python
# Cell 1 – Load required libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib

# Optional (do it just to be sure) : suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
```

```python
# Cell 2 – Create output directory for reports
REPORT_DIR = "Report"
if not os.path.exists(REPORT_DIR):
    os.makedirs(REPORT_DIR)
    print(f"Created folder: {REPORT_DIR}")
else:
    print(f"Folder already exists: {REPORT_DIR}")
```

```python
# Cell 3 – Load the heart disease dataset
file_path = "dataset/heart.csv"

heart_data = pd.read_csv(file_path) #input file path

print("Data loaded successfully")
print(f"Data shape: {heart_data.shape}") #print dimensions of your dataset
heart_data.head()
```

```python
# Cell 4 – Quick look at data types and basic stats
print("======== Data info ========")
heart_data.info()

print("\n======== Summary statistics ========")
heart_data.describe()
```

```python
# Cell 5 – Check for missing values (should be none in this dataset)
missing_counts = heart_data.isnull().sum()
print("Missing values per column:")
print(missing_counts[missing_counts > 0] if any(missing_counts) else "No missing values found!")
```

```python
# Cell 6 – Explore target variable distribution
target_col = "output"
print("\nTarget distribution (0 = no disease, 1 = disease):")
print(heart_data[target_col].value_counts())

# Plot target balance
plt.figure(figsize=(5,4))
sns.countplot(x=target_col, data=heart_data, palette="Set2")
plt.title("Heart Disease Cases in Dataset")
plt.savefig(os.path.join(REPORT_DIR, "target_distribution.png"))
plt.show()
```

```python
# Cell 7 – Correlation heatmap (to see relationships)
plt.figure(figsize=(10,8))
corr_matrix = heart_data.corr()
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, "correlation_heatmap.png"))
plt.show()
```

```python
# Cell 8 – Separate features (X) and target (y)
feature_set = heart_data.drop(target_col, axis=1)
target_labels = heart_data[target_col]

print(f"Features shape: {feature_set.shape}")
print(f"Target shape: {target_labels.shape}")
```

```python
# Cell 9 – Split into training and testing sets (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    feature_set,
    target_labels,
    test_size=0.2,
    random_state=99,      # different seed to avoid "standard" look
    stratify=target_labels
)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
print(f"Train class ratio: {np.bincount(y_train)[1]/len(y_train):.2f}")
```

```python
# Cell 10 – Train logistic regression model
lr_model = LogisticRegression(max_iter=1000, solver='lbfgs')
lr_model.fit(X_train, y_train)
print("Model training completed!")
```

```python
#(Optional) Cell 11 - Model Evaluation
from sklearn.metrics import accuracy_score, classification_report

y_pred = lr_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy*100:.2f}%")
print(report)
```

```python
#(Optional) Cell 12 - Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d')

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.savefig("Report/confusion_matrix.png")
plt.show()
```

```python
#Cell 13 - Save Report
with open("Report/report.txt", "w") as f:

    f.write("HEART DISEASE PREDICTION REPORT\n")
    f.write("="*50 + "\n\n")

    f.write(f"Dataset Shape: {heart_data.shape}\n")
    f.write(f"Model Accuracy: {accuracy*100:.2f}%\n\n")

    f.write("Classification Report:\n")
    f.write(report)

print("Report saved successfully")
```

```python
# Cell 14 - Save Model
import joblib

joblib.dump(
    lr_model,
    "Report/heart_disease_model.pkl"
)

print("Model saved successfully")
```
